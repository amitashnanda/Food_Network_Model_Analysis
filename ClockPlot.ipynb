```python
pd.set_option('display.max_rows', 100)                                #setting the column
df = pd.read_excel("HealthDataSet.xlsx", header=None).iloc[1:]

df.columns =['ID', 'Date', 'Time', 'Food_Entry', 'Food_Type','Process_Food', 'Value']
df.head(100)
```

```python
count = final_data[['Date','Time','Process_Food']].astype(str)
count.head(100)
```

```python

import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure(figsize=(8,8))
ax = plt.subplot(111, projection = 'polar', facecolor="lightgoldenrodyellow")


equals = np.linspace(0, 360, 24, endpoint=False, axis=0) #np.arange(24)
ones = np.ones(24)
sizes = np.array(list(final_dict1.values()))*50
colors = np.array(list(final_dict1.values()))*50


ax.scatter(np.deg2rad(equals), ones,  c=colors, s=sizes)


# r = np.linspace(0, 3, 301)
# theta = 2 * np.pi * r
# ax.plot(theta, r, color="tab:orange", lw=3, label="a line")
# ax.plot(0.5 * theta, r, color="tab:blue", ls="--", lw=3, label="another line")
ax.tick_params(grid_color="palegoldenrod", grid_alpha=0.4)
# ax.tick_params(grid_color="palegoldenrod")



# Make the labels go clockwise
ax.set_theta_direction(-1)

#Place Zero at Top
ax.set_theta_offset(np.pi/2)


theta1 = np.array([np.pi / 2] * 100)
r2 = np.linspace(0, 2, 100)
ax.plot(theta1, r2, color='k', alpha=0.01)

theta2 = np.array([np.pi / 1] * 100)
r2 = np.linspace(0, 2, 100)
ax.plot(theta2, r2, color='k', alpha=0.01)

theta3 = np.array([2*np.pi / -12] * 100)
r2 = np.linspace(0, 2, 100)
ax.plot(theta3, r2, color='k', alpha=0.01)

theta4 = np.array([2*np.pi /-3 ] * 100)
r2 = np.linspace(0, 2, 100)
ax.plot(theta4, r2, color='k', alpha=0.0)

theta5 = np.linspace(2*np.pi/-12, 2*np.pi/4, 100)
ax.fill_between(theta5, 1.5, 2, color='deepskyblue',  label="Sleep")

theta6 = np.linspace(2*np.pi/4,2*np.pi/2, 100)
ax.fill_between(theta6, 1.5, 2, color='purple',  label="Breakfast")

theta7 = np.linspace(2*np.pi/2,2*np.pi/1.5, 100)
ax.fill_between(theta7, 1.5, 2, color='gold',  label="Lunch")

theta8 = np.linspace(2*np.pi/1.5,2*np.pi/1.09, 100)
ax.fill_between(theta8, 1.5, 2, color='red',  label="Dinner")



#Set the circumference ticks
ax.set_xticks(np.linspace(0, 2*np.pi, 24, endpoint=False, ))

# set the label names
ticks = ['12 AM', '1 AM', '2 AM', '3 AM', '4 AM', '5 AM', '6 AM', '7 AM','8 AM','9 AM','10 AM','11 AM','12 PM', '1 PM', '2 PM', '3 PM', '4 PM',  '5 PM', '6 PM', '7 PM', '8 PM', '9 PM', '10 PM', '11 PM' ]
ax.set_xticklabels(ticks)

# suppress the radial labels
plt.setp(ax.get_yticklabels(), visible=False)

#Bars to the wall
plt.ylim(0,2)

# plt.legend(bbox_to_anchor=(1,0), fancybox=True, shadow=True)

# cursor = mplcursors.cursor(ax, hover=True)
# @cursor.connect("add")
# def on_add(sel):
#     sel.annotation.set(text=tt[sel.target.index])
angle = np.deg2rad(67.5)
ax.legend(loc="lower left",
          bbox_to_anchor=(.7 + np.cos(angle)/2, .5 + np.sin(angle)/2))

    

plt.show()
fig.savefig('clockplot1.png')
```
